```python
import requests
import json
import bs4 as beautifulsoup
import csv
```

```python
url= "https://www.1mg.com/"
```

```python
r= requests.get(url)
htmlContent= r.content
print(htmlContent)
```

```python
soup= beautifulsoup.BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify)
```

```python
title= soup.title
print(title)
```

```python
paras= soup.find_all('p')
print(paras)
```

```python
scripts= soup.find_all('script')
print(scripts)
```

```python
for script in scripts:
    print(script.get_text())
```

```python
# Get all the links on the page
anchors= soup.find_all('a')
all_links= set()
for link in anchors:
    if(link.get('href')!= '#'):
        linkText= "https://www.1mg.com/"+link.get('href')
        all_links.add(link)
        print(linkText)
```

```python
soup= beautifulsoup.BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify)
```

```python
medicines = soup.find_all('div', class_='style__container___1YjHN')
print(medicines)
```

```python
medicines = soup.find_all('div', class_='style__pro-title___3BTz6')
print(medicines)
```

```python
with open('medicines.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'composition', 'effectiveness'])
    for medicine in medicines:
        name = medicine.find('div', class_='style__pro-title___3BTz6').text
        composition = medicine.find('div', class_='style__salt-info___1ZyNG').text
        effectiveness = medicine.find('div', class_='style__strength___1MaKg').text
        writer.writerow([name, composition, effectiveness])
```
